```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
import statistics


browser = webdriver.Chrome()
browser.get("https://m.imdb.com/chart/top/")
time.sleep(5)
browser.maximize_window()
time.sleep(3)

## XPATH'leri liste olarak tanımlıyoruz.
xpaths = ["//*[@id='chart-content']/div[{}]/div[{}]/div/a/span".format(i, j) for i in range(1, 126) for j in range(1, 3)]

## Her bir öğeyi seçiyoruz.
for xpath in xpaths:
    elm = browser.find_element(By.XPATH, xpath)
    print(elm.text)
    
```

```python
## Sadece 'rating'lerin XPATH'lerini liste olarak tanımlıyoruz.
xpaths1 = ["//*[@id='chart-content']/div[{}]/div[{}]/div/a/span/p/span[2]".format(i,j) for i in range(1,126) for j in range(1,3)]

puanlar_list = []

## Her bir öğeyi seçiyoruz ve listeye ekliyoruz.
for xpath1 in xpaths1:
    ratings = browser.find_element(By.XPATH, xpath1)
    puanlar_list.append(ratings.text)
    
print(puanlar_list)  

puanlar_list = [float(ratings.replace(',', '.'))for ratings in puanlar_list]

print("----------HESAPLAMALAR----------")

print(f"Mod: {statistics.mode(puanlar_list)}")
print(f"Medyan: {statistics.median(puanlar_list)}")
print(f"Ortalama: {statistics.mean(puanlar_list)}")
print(f"En Yüksek Puan: {max(puanlar_list)}")
print(f"En Düşük Puan: {min(puanlar_list)}")
print(f"Standart Sapma: {statistics.stdev(puanlar_list)}")
```
